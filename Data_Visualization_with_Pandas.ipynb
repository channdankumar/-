```python
import pandas as pd

# Try reading the CSV file with 'latin1' encoding
df = pd.read_csv('/content/cars_dataset.csv', encoding='latin1')

# Now you can work with the DataFrame 'df'
```

##Plot()
- It makes charts - default is line plot
- More graphs with kind='line','bar', 'scatter', 'hist' ,etc

##Line Plot
syntex: plot() or plot.line()

```python
df.plot()
```

## Scatter Plot

syntex:plot(kind='scatter', x,y) or plot.scatter(x,y)

```python
df.plot.scatter(x='year', y='tax')
```

##Histrogram

```python
df.plot(kind='hist')
```

##Area Plot
syntex : plot(kind='area') or plot.area()

```python
df.plot.area()
```
